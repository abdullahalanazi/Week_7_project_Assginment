```python
#importing smote to oversample the minority class to balance the class distribution

from imblearn.over_sampling import SMOTE

x_up, y_up = SMOTE().fit_resample(X,Y)
pca1 = pca()
X_scaled = pca1.fit_transform(x_up)
index = int (len(X_scaled)*0.7)
X_train = X_scaled[:index]
y_train = Y[:index]
y_test = X_scaled[index:]
X_test = Y[index:]
X_scaled
```

```python
#split into train and test the set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.5)
```

```python
len(X_test)
```

```python
#fit model with training data
model= RandomForestClassifier(n_estimators=60)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
#calculating errors using test data (cross validation)
print(classification_report(y_test,y_pred))
```

```python
print(model.score(X_test,y_test))
```
